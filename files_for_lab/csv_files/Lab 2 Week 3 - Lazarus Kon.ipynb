Linear Regression

   * Train-test split.
   * Apply linear regression.


```python
from sklearn.model_selection import train_test_split #applying train-test split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=0)
```

```python
X_test
```

```python
X_train #sanity checks
```

```python
from sklearn.preprocessing import StandardScaler #normalizing the explantory train+test variables
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
X_train #sanity check to see if they turned into arrays
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression #applying linear regression
regressor = LinearRegression() 
regressor.fit(X_train, y_train)
```

```python
regressor.coef_ #checking the coeffecients of the regressor
```

Model Validation

   * Description:
       * R2.
       * MSE.
       * RMSE.
       * MAE.


```python
#to get a general idea of how our model performs:
```

```python
print(" Score of Train : {}\n Score of Test : {}".format(regressor.score(X_train, y_train), regressor.score(X_test, y_test)))
```

```python
#We see a huge difference between train and score here. 
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score


y_pred = regressor.predict(X)

r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred, squared=True)
rmse = mean_squared_error(y, y_pred, squared=False)
mae = mean_absolute_error(y, y_pred)

print("R2_score:", r2)
print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
```
